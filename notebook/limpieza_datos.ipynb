```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
import os

# Crear carpeta data si no existe
os.makedirs("data", exist_ok=True)
```

```python
df = pd.read_csv('../data/creditcard.csv', sep=';', low_memory=False)

# Ver datos
df.head()
```

```python
# Ver nombres de columnas
print(df.columns)

# Quitar espacios por si acaso
df.columns = df.columns.str.strip()
```

```python
for col in df.columns:
    if col != 'Class':
        df[col] = df[col].astype(str).str.replace('.', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.isnull().sum()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

print(X.shape, y.shape)
```

```python
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Comprobar que no hay NaN
print(np.isnan(X).sum())
```

```python
columnas = df.drop('Class', axis=1).columns

X_df = pd.DataFrame(X, columns=columnas)

y_df = pd.DataFrame(y, columns=['Class'])

df_clean = pd.concat([X_df, y_df], axis=1)

df_clean.head()
```

```python
df_clean.isnull().sum().sum()
```

```python
df_clean.to_csv("data/creditcard_limpio.csv", index=False)
```
